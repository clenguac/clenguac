```python
# Cargar datos con phyton

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Carga de Datos: Usar Pandas para cargar los datos desde un archivo CSV.
from google.colab import drive
drive.mount('/gdrive')

# Reemplaza 'nombre_del_archivo.csv' con la ruta real de tu archivo CSV
data = pd.read_csv('/gdrive/MyDrive/Colab Notebooks/estadistica/titanic.csv')
```

# Estadisticas descriptivas


```python
# Mostrar las primeras filas del DataFrame
print(data.head())

# Obtener estadísticas descriptivas del DataFrame
print(data.describe())

# Visualizar el DataFrame en una tabla
data
```

```python

# Estadísticas descriptivas
print(data.describe())
```

**Visualizar columna**

```python
# Visualizar la columna 'Edad' con un histograma
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], kde=True)  # kde=True agrega una curva de densidad
plt.title('Distribución de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Calcula la media de la edad y redondea a dos decimales
media_edad = data['Age'].mean()
media_edad_rounded = round(media_edad, 1)

# Imprime la media de la edad redondeada
print(f"La media de la edad es: {media_edad_rounded}")
```

```python
# Calcula la moda de la edad y redondea a un dígito
moda_edad = data['Age'].mode()

# Si hay múltiples modas, toma la primera
if isinstance(moda_edad, pd.Series):
  moda_edad = moda_edad.iloc[0]

# Redondea la moda a un decimal
moda_edad_rounded = round(moda_edad, 1)

# Imprime la moda de la edad redondeada
print(f"La moda de la edad es: {moda_edad_rounded}")
```

# **Correlación**

```python
# Calcula la correlación entre 'Edad' y 'Precio del Tiquete'
correlation = data['Age'].corr(data['Fare'])

# Imprime la correlación
print(f"La correlación entre la edad y el precio del tiquete es: {correlation}")

# Crea un gráfico de dispersión para visualizar la relación entre 'Age' y 'Fare'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', data=data)
plt.title('Relación entre Edad y Precio del Tiquete')
plt.xlabel('Edad')
plt.ylabel('Precio del Tiquete')
plt.show()
```
